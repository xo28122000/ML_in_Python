```python
import csv

data = []

with open('data/Wine_Quality/winequality-white.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=';')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            head = row
            line_count += 1
        else:
            data.append(row)
            line_count += 1
    print(f'Processed {line_count} lines.')
```
