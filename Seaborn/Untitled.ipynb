## THE IRIS DATASET

```python
from sklearn.datasets import load_iris
data = load_iris()

X = data.data
Y = data.target
```

```python
iris_dataset = pd.DataFrame({"sepal length": X[:, 0], "sepal width": X[:, 1],"petal length": X[:, 2],"petal width": X[:, 3],"target": Y})
```

```python
sns.relplot(x ="petal length",  y = "target", hue = "petal width",data = iris_dataset );
```

```python
sns.pairplot(iris_dataset)
```

### 1) pima_indians_diebetes
../Mini\ Projects/data/pima_indians_diebetes/pima-indians-diabetes.data.csv

```python
import csv
```

```python
pima_rawdata = []
with open("../Mini Projects/data/pima_indians_diebetes/pima-indians-diabetes.data.csv") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
            pima_rawdata.append(row)
            line_count += 1
    print(f'Processed {line_count} lines.')
```

```python
pima_rawdata = np.array(pima_rawdata)
np.shape(pima_rawdata)
```

```python
pima_dataset = pd.DataFrame({"Number of times pregnant":pima_rawdata[:,0], "Plasma glucose concentration":pima_rawdata[:,1],"Diastolic blood pressure":pima_rawdata[:,2],"Triceps skinfold thickness":pima_rawdata[:,3],"serum insulin" :pima_rawdata[:,4], "Body mass index":pima_rawdata[:,5],"Diabetes pedigree function":pima_rawdata[:,6],"Age" :pima_rawdata[:,7], "target":pima_rawdata[:,8] })
```

```python
pima_dataset_1 = (pima_dataset[["Number of times pregnant","Plasma glucose concentration","Diastolic blood pressure","Triceps skinfold thickness","target"]])
pima_dataset_2 = (pima_dataset[["serum insulin","Body mass index","Diabetes pedigree function","Age","target"]])
```
